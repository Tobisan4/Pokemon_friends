```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
### Store CSV into DataFrame
```

```python
# Store Pokemon filepath in a variable
pokemon_file = "Resources/PokemonDb.csv"
pokemon_data_df = pd.read_csv(pokemon_file)
# Show just the header
pokemon_data_df.head()
```

```python
#Store Chick-Fil-A filepath in a variable
chik_fil_a_file = "Resources/chick-fil-a-nutrition.csv"
chik_fil_a_data_df = pd.read_csv(chik_fil_a_file)
# Show just the header
chik_fil_a_data_df.head()
```

```python
# Store Chick-fil-a filepath in a variable
chick_fil_a_file = "Resources/chick-fil-a-nutrition.csv"
chick_fil_a_data_df = pd.read_csv(chick_fil_a_file)
# Show just the header
chick_fil_a_data_df.head()
```

```python
# Filter for only the basic types of Pokemons
pokemon_data_df = pokemon_data_df.loc[pokemon_data_df["Variation"].isna(), :]
pokemon_data_df
```

```python
# Drop columns that are not needed
pokemon_data_df = pokemon_data_df.drop(columns = ["Variation", "Total"])
pokemon_data_df
```

```python
# Replace NaN values with a space
pokemon_data_df = pokemon_data_df.fillna("")
```

```python
# Sort Pokemons first by Type1 then by Type2
pokemon_data_df = pokemon_data_df.sort_values(by = ['Type1', 'Type2'], ascending = [True, True])
pokemon_data_df
```

```python
# Calculate the Total Offensive and Total Defensive stats
for name in pokemon_data_df["Name"]:
    pokemon_data_df["Total Attack"] = pokemon_data_df["Attack"] + pokemon_data_df["Sp.Atk"]
    pokemon_data_df["Total Defense"] = pokemon_data_df["Defense"] + pokemon_data_df["Sp.Def"]
pokemon_data_df
```

```python
# Capitalize the first letter of each Pokemon Name, Type1 and Type2
pokemon_data_df["Name"] = pokemon_data_df["Name"].str.capitalize()
pokemon_data_df["Type1"] = pokemon_data_df["Type1"].str.capitalize()
pokemon_data_df["Type2"] = pokemon_data_df["Type2"].str.capitalize()
pokemon_data_df
```

```python
# Reset index
pokemon_data_df.reset_index(inplace = True, drop = True)
pokemon_data_df
```

```python
# Filter for single serving items
chik_fil_a_data_df = chik_fil_a_data_df[chik_fil_a_data_df["Menu"].str.contains("Tray|Gallon|Bucket|8oz|Ice Scoop|96|Catering|Bag")==False]
chik_fil_a_data_df
```

```python
# Edit "Serving Size" column, remove "g" and change type to float
chik_fil_a_data_df['Serving size'] = chik_fil_a_data_df['Serving size'].str.replace('g','')
chik_fil_a_data_df['Serving size'] = pd.to_numeric(chik_fil_a_data_df['Serving size'], downcast='float')
chik_fil_a_data_df.head()
```

```python
# Rename 'Serving size' to:'Serving Size (G)'
chik_fil_a_data_df.rename(columns = {'Serving size':'Serving Size (G)'}, inplace = True)
chik_fil_a_data_df
```

```python
# Sorting data by "Calories" and "Serving Size (G)"
chik_fil_a_data_df = chik_fil_a_data_df.sort_values(by = ['Calories', 'Serving Size (G)'], ascending = [False, False])
chik_fil_a_data_df.head()
```

```python
# Reset index
chik_fil_a_data_df.reset_index(inplace=True, drop=True)
chik_fil_a_data_df.dtypes
```

```python
# Connect to the database
rds_connection_string = "postgres:postgres@localhost:5432/Pokemon_friends"
engine = create_engine(f"postgresql://{rds_connection_string}")
```

```python
# Use pandas to load csv converted to dataFrame into database (Chick-fil-a)
chik_fil_a_data_df.to_sql(name = "chick_fil_a", con = engine, if_exists = "append", index=False)
```
